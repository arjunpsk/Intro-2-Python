```python
# Exercise 7 (while loop):
age_DA = [32, 37, 30, 29, 100, 57, 64]
age_i = 0
print("The following is the entire list of ages:")
while age_i < len(age_DA):
    print(age_DA[age_i])
    age_i = age_i + 1
print("")
age_even = 0
print("You'll understand the following because it says what the numbers are:")
while age_even < len(age_DA):
    if age_DA[age_even] % 2 == 0:
        print(age_DA[age_even], "is even")
    age_even = age_even + 1
```

```python
# Exercise 19 - Part 02
# This is where the user is asked to enter a list of ages. 
# The only accepted input is integer. 
# If the value is not an integer an error "Invalid input" is shown.
# When the user is finished inputing the age list, the user must type "done" to finish the age list
# The user is then asked to enter a number to check with the function defined in Exercise 19 - Part 01. 
age_input = []
active=True
while active:
    sval = input("Enter ages to create a list of ages. (Note: please only enter integers.) Type done when you are finished: ")
    if sval == 'done' :
        active=False
    try:
        age_input.append(int(sval))
    except:
        print('Invalid input')
        continue
print(age_input)

# This is where the age to check with the list is entered. 
sval_01 = int(input("Please enter one age:"))
check_age(sval_01)

print("The age entered to check is", sval_01)
```
